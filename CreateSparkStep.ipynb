# Create Spark Step
___

### Step 1.

1. 建立  [Amazon](https://aws.amazon.com/free/) 會員。會員部分要注意，在地址部分，只能填上英文。
2. 注意信用卡問題，Amazon 加入會員須要預扣 ***1元美金*** ，如果沒有預扣成功，將無法使用 Amazon 上面所有的服務。部分銀行會阻擋預扣，當遇扣失敗時， Amazon 會發信，請跟信用卡銀行進行聯絡，並開通預扣這部分，等到預扣成功，就可以使用 Amazon 上的服務。
___

### Step 2.

1. 等到會員成功後，即可使用任何服務，點進 ***服務*** 後，點選運算內的 ***EC2*** 。


2. 點進去後，會看到 ***Create Instance*** ，下面有一個 ***Launch Instance***，點進去。


3. 點進去後，本次安裝使用 Ubuntu Server，下面有標記 ***Free for Eligible***，有標記這個才是免費使用，否則會被另外索取費用。


4. 在後面的步驟，都使用預設的，不需要更改任何設定。


5. 直到 ***Step 5: Tag Instance*** ，需要自己建立 ***Create Tag***，並輸入自己想的 key 和 value。


6. 在 ***Step 6: Configure Security Group*** ，找到 ***Assign a security group*** 地方，點選 ***Create a new security group***。


7. 其餘步驟皆是預設的，直到看到 ***Select an existing key pair or create a new key pair***，這裡要點選 ***Create a new key pair***，並輸入 ***key pair name***，然後點選 ***download key pair***，會自動下載 pem 檔案，下載後要保留好，一旦該步驟結束後，就無法再度下載。


8. 以上步驟都好了之後， instance 就建立好，就會看到 instance pending，pending 好之後，就會出現 running 的樣子，代表 instance 建立成功。
___

### Step 3.

***此步驟適用於 Mac or Linux ，Windows 請另外尋找建立 SSL 的步驟。***

1. 打開 terminal，<code>cd Desktop</code>。


2. 輸入指令 <code>chmod 400 xxx.pem</code>。


    2-1. 這個 pem 就是你在上面步驟 7 所下載的 pem 檔案，並確保這個檔案在桌面。
    2-2. xxx 代表這個 pem 的檔名。

3. 輸入指令：<code>ssh -i xxx.pem ubuntu@xxxxxxxxxxxxxx.us-east-2.compute.amazonaws.com</code>。


    3-1. xxx.pem，就是所下載的 pem 檔名。
    3-2. @後面的，就是所建立 instance Public DNS，全部複製，並貼上去。
    
4. 接著就在畫面出現 continue connecting 的地方，輸入 yes。


5. 最後看到 ubuntu@xxxxx 就代表成功了。
___

### Step 4.

***在上面安裝 Anaconda ，並順利執行 Jupyter notebook。***

1. 進去 ubuntu 之後，輸入指令 <code>wget http://repo.continuum.io/archive/Anaconda3-4.1.1-Linux-x86_64.sh</code>

    
    1-1. 可在 url 地方輸入，Anaconda版本(http://repo.continuum.io/archive/) 選擇要安裝的版本，要選擇 Linux-x86_64.sh 結尾。
    

2. 安裝之後，輸入指令 <code>bash Anaconda3–4.1.1-Linux-x86_64.sh</code>，執行後，就會看到 ***please review the license***，一路按 enter，直到看到 ***Please Answer yes or no***，輸入 yes，之後會看到 **[/home/ubuntu/anaconda3]**，按 enter，就會開始進行安裝。


3. 之後，輸入指令<code>which python</code>，會看到 **[usr/bin/python]**，但要看到在 anaconda 內的 python，所以再度輸入<code>source .bashrc</code>，在輸入<code>which python</code>，看到 **[/home/ubuntu/anaconda3/bin/python]** 就對了。


4. 接下來要 configure jupyter notebook，輸入 <code>jupyter notebook --generate-config</code>，輸入後會看到 ***Writing default config to :/home/ubuntu/.jupyter/jupyter_notebook_config.py***，看到顯示出這個，就成功了。


5. 接著要建立 jupyter notebook 的 SSL 連結，所以要建立一個存放 cert 的地方，因此輸入指令 <code>mkdir certs</code>，並進入這個資料夾，輸入 <code>cd certs</code>，進入 certs 後，在輸入<code> sudo openssl req -x509 -nodes -days 365 -newkey rsa:1024 -keyout mycert.pem -out mycert.pem</code>，接著畫面就會出現： ***Country Name***，一定要依序跟著顯示出的指令往下輸入，因為此步驟在製作憑證，如若不輸入，則後面憑證地方會出現錯誤。


6. 憑證完成之後，輸入 <code>cd ~/.jupyter/</code>進入這個資料夾，並在輸入<code>vi jupyter_notebook_config.py</code>，會看到visual editor，看到之後，移到最上面，輸入 i（代表 insert），進行編輯。


    6-1. 在編輯模式下，請輸入以下程式碼：（註解可以不用輸入）
        
        c = get_config()
        # Notebook config this is where you saved your pem cert
        c.NotebookApp.certfile = u'/home/ubuntu/certs/mycert.pem' 
        # Run on all IP addresses of your instance
        c.NotebookApp.ip = '*'
        # Don't open browser by default
        c.NotebookApp.open_browser = False  
        # Fix port to 8888
        c.NotebookApp.port = 8888
        
    6-2. 輸入完畢後，按 ESC，接著按 :wq（w 代表 write, q 代表 quite） 
    

7. 接著就執行 jupyter notebook，輸入指令 <code>jupyter notebook</code>，並去 AWS 複製 public DNS，並去網址列輸入 <code>https://xxxxxxxxxxxx.us-east-2.compute.amazonaws.com:8888</code>，就會進入 jupyter notebook。



8. 以上都好後，就安裝所需套件，第一個先安裝 java，輸入指令<code>sudo apt-get update</code>，緊接著在輸入<code>sudo apt-get install default-jre</code>，安裝完後，輸入<code>java -version</code>，就可以看到 java 版本。


9. 接著安裝 scala，輸入指令<code>sudo apt-get install scala</code>，緊接著在輸入<code>scala -version</code>，就可以看到 scala 版本。


10. 最後安裝 spark python 所需套件py4j，輸入指令<code>export PATH=$PATH:$HOME/anaconda3/bin</code>，緊接著在輸入<code>conda install pip</code>，最後輸入<code>pip install py4j</code>。


11. 最後要安裝 Hadoop 和 Spark，第一個要先輸入<code>wget http://archive.apache.org/dist/spark/spark-2.0.0/spark-2.0.0-bin-hadoop2.7.tgz</code>。


    11-1 可在 url 地方輸入，Spark版本(http://archive.apache.org/dist/spark/)選擇要安裝的版本，選擇 spark-2.4.4-bin-hadoop2.7.tgz。


12. 安裝好後，進行解壓縮，輸入指令<code>sudo tar -zxvf spark-2.4.4-bin-hadoop2.7.tgz</code>。


13. 為了讓 Anaconda 找到 Spark，需要將路徑指到 Spark。


    13-1. 首先輸入指令：export SPARK_HOME='/home/ubuntu/spark-2.4.4-bin-hadoop2.7'。
    13-2. 接著在輸入：export PATH=$SPARK_HOME:$PATH。
    13-3. 最後輸入：export PYTHONPATH=$SPARK_HOME/python:$PYTHONPATH。


14. 最後在執行第七個步驟即可執行 jupyter notebook，執行完後，新增檔案，並輸入<code>from pyspark import SparkContext</code>，在輸入<code>sc = SparkContext()</code>，如若能夠順利執行，則代表安裝成功。
